```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

```python
h = ['Date', 'Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustDir',
    'WindGustSpeed', 'WindDir9am', 'WindDir3pm', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
    'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RainToday', 
     'Risk_MM', 'RainTomorrow']
df = pd.read_csv('/Users/simrankodwani/Downloads/weatherPerth.csv', names=h)
```

```python
# exclude = ['Date', 'RISK_MM', 'Location']
df = df.drop(columns=['Date', 'Risk_MM', 'Location'])
```

```python
df = df.dropna()
```

```python
df
```

```python
df['RainToday'] = np.where(df['RainToday'] == 'Yes', 1, 0)
df['RainTomorrow'] = np.where(df['RainTomorrow'] == 'Yes', 1, 0)
```

```python
df
```

```python
directions = ['N', 'NNE', 'NE', 'ENE', 'E', 'ESE', 'SE', 'SSE', 'S', 'SSW', 'SW', 'WSW', 'W', 'WNW','NW', 'NNW']
angles = np.arange(0.0, 2.0 * np.pi, 2.0 * np.pi/16.0)
wind_angles = dict(zip(directions, angles))
print(wind_angles)
```

```python
wind_attributes = ['WindGustDir', 'WindDir9am', 'WindDir3pm']
for att in wind_attributes:
    df[att] = df[att].map(wind_angles)
    df[att + '_cos'] = np.cos(df[att])
    df[att + '_sin'] = np.sin(df[att])
    df = df.drop(columns=att)
```

```python
df
```

```python
y = df['RainTomorrow']
X = df.drop(columns=['RainTomorrow'])
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
mn = MLPClassifier(hidden_layer_sizes=[50,50], random_state=8,
                  max_iter=500)
mn.fit(X_train, y_train)
```

```python
y_pred = mn.predict(X_test)
```

```python
r = accuracy_score(y_test, y_pred)
print('The accuracy of the model is:',r)
```

```python
p = {'hidden_layer_sizes': ((2,), (10,), (50,), (75,), (50,50), (75,75), (100,100))}
mn1 = MLPClassifier(max_iter=2000, random_state=0)
gs = GridSearchCV(mn1, p, cv=5)
gs.fit(X_train, y_train)
```

```python
gs.cv_results_['mean_test_score']
```

```python
best_mn = gs.best_estimator_
y_predmn = best_mn.predict(X_test)
```

```python
accuracy_score(y_test, y_predmn)
```
